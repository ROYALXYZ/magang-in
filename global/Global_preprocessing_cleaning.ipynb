# Global Dataset Preprocessing & Cleaning

## Objective
Menyiapkan dataset global agar:
- Bersih (clean)
- Konsisten
- Memiliki struktur yang sama dengan dataset local
- Siap digunakan untuk EDA dan comparison

---

## 1. Load Dataset
- Import library (`pandas`, `numpy`)
- Load dataset dari Kaggle
- Cek shape dan preview data (`head()`, `info()`)

---

## 2. Handle Missing Values
- Identifikasi kolom yang memiliki null
- Tentukan strategi:
  - Drop jika sedikit
  - Impute jika diperlukan
- Pastikan kolom penting tidak kosong:
  - `skills`
  - `job_title`
  - `role`

---

## 3. Filter Data (Tech Only)
- Pastikan hanya data tech-related jobs
- Gunakan keyword filtering atau rule-based:
  - contoh: data, developer, engineer, AI, dll
- Buat kolom:
  - `is_tech` → True/False
- Filter hanya `is_tech = True`

---

## 4. Standardisasi Kolom
Samakan dengan dataset local:

- Rename kolom agar konsisten:
  - `job_title` → `title`
  - `company` → `company_name`
- Pastikan kolom berikut tersedia:
  - `title`
  - `company_name`
  - `skills`
  - `role`
  - `location_city`

---

## 5. Cleaning & Normalisasi Skills
- Lowercase semua skill
- Hapus karakter aneh
- Pisahkan skill menjadi format konsisten (comma-separated atau list)
- Simpan ke kolom:
  - `skills_cleaned`

---

## 6. Feature Engineering
Tambahkan fitur yang sama dengan local dataset:

- `skills_count`
  - jumlah skill per job
- `location_city`
  - ekstrak dari location (jika perlu)
- `role`
  - mapping job title → role category

---

## 7. Final Check
- Cek:
  - Missing values
  - Duplicate data
  - Konsistensi format
- Pastikan schema sama dengan dataset local

---

## 8. Save Clean Dataset
- Simpan sebagai:
  - `global_jobs_cleaned.csv`
- Pastikan siap digunakan untuk EDA
